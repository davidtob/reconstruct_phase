```python
%pylab inline
from pylab import plt
import sys
import scipy
import audiodisplay
sys.path.insert(0,'../timitdataset')

from timit_phase_from_amp import TIMITPhaseFromAmp

v = TIMITPhaseFromAmp('valid', stop=1)
```

```python
def add_phase_noise(kappa):
    x = numpy.linspace(-numpy.pi,numpy.pi,100)
    plt.figure( figsize=(10,2.5) )
    plt.subplot( 1, 2, 1 )
    if kappa>0:
        plt.plot( x, scipy.stats.distributions.vonmises.pdf(x,kappa ) )
    else:
        plt.plot( x, x*0 + 1/(2*numpy.pi) )
    plt.ylim( [0,5] )
    plt.title('PDF of angle of noise (von Mises density)')
    plt.xlabel( 'angle (radians)' )
    plt.ylabel( 'density' )
    
    complex_phases = v.y[0:100,:120] + 1j*v.y[0:100,120:]
    if kappa>0:
        complex_noise = numpy.exp( 1j * numpy.random.vonmises( 0, kappa, (100,120) ) )
    else:
        complex_noise = numpy.exp( 1j * numpy.random.uniform( -numpy.pi, numpy.pi, (100,120) ) )
    new_complex_phases = complex_phases * complex_noise
    new_real_phases = numpy.hstack( (numpy.real(new_complex_phases),numpy.imag(new_complex_phases)) )
    
    new_sound = v.to_audio( v.X[0:100,:], new_real_phases )
    plt.subplot( 1, 2, 2 )
    plt.plot(new_sound/542.0)
    plt.title( 'Sound with noise added' )
    
    print "MRSE", numpy.sqrt(numpy.mean((new_sound/542.0-originalsound/542.0)**2))
    return audiodisplay.Audio( rate=16000, data=new_sound.astype('int16') )
```
